```python
import matplotlib.pyplot as plt

# Sample data
genders = ['Male', 'Female', 'Non-binary', 'Prefer not to say']
counts = [300, 350, 20, 30]

# Create bar chart
plt.bar(genders, counts)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders in the Population')
plt.show()
```

```python
import seaborn as sns

# Sample data
ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]

# Create histogram
sns.histplot(ages, kde=False)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Ages in the Population')
plt.show()
```
